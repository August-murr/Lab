```python
from pymongo.mongo_client import MongoClient
from pymongo.server_api import ServerApi
from kaggle_secrets import UserSecretsClient
```

```python
# Connect to MongoDB
user_secrets = UserSecretsClient() #Replace with your MongoDB connection string
mongo_url = user_secrets.get_secret("mongodb")
client = MongoClient(mongo_url, server_api=ServerApi('1')) 
```

```python
db = client['gsm8k_dataset']  # Assuming the questions dataset is in your MongoDB
```

```python
# Collections
gsm8k_collection = db['gsm8k_train']  # Existing collection with questions
run_collection = db['base_mistral_train_run']  # New collection for the run
```

```python
# Fetch all questions from the GSM8k collection
questions = gsm8k_collection.find({}, {'_id': 0, 'question': 1})  # Fetch only the 'question' field
```

```python
# Prepare documents for the new collection
run_documents = []
for question in questions:
    run_documents.append({
        "question": question['question'],
        "response": "",
        "instance": "",
        "status": "pending"
    })
```

```python
# Insert documents into the new collection
try:
    run_collection.insert_many(run_documents)
    print("Run table created and data inserted successfully!")
except Exception as e:
    print(f"An error occurred while creating the run table: {e}")
```
